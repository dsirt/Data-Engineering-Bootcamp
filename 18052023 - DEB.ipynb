```python
plt.figure(figsize = (6,6))
area = [48, 30, 20, 15]
total = np.sum(area)
labels = ['Low', 'Medium', 'High', 'Critical']

def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x, total*x/100)

colors = ['#7CB342', '#C0CA33', '#FFB300', '#F57C00']
plt.pie(area, labels = labels, colors = colors, startangle = 45, pctdistance = 0.85, autopct = val_per)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf() #get current figure
p.gca().add_artist(my_circle) #get current axes
plt.show()
```

```python
#Multiple Donut Plot
```

```python
fig = plt.figure(figsize = (20, 6))
area = [48, 30, 20, 15]
priority = ['Low', 'Medium', 'High', 'Critical']
status = ['Resolved', 'Cancelled', 'Pending', 'Assigned']
company = ['Tesla', 'Microsoft', 'Apple', 'TOGG']
colors = ['#7BC342', '#C0CA33', '#FFB300', '#F57C00']

plt.subplot(1,3,1)
plt.pie(area, labels = priority, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.subplot(1,3,2)
plt.pie(area, labels = status, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.subplot(1,3,3)
plt.pie(area, labels = company, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
x = np.random.normal(size = 2000)
plt.hist(x, bins = 40, color = 'yellowgreen')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
plt.figure(figsize = (10,8))
x = np.random.normal(size = 2000)
plt.hist(x, bins = 40, color = 'yellowgreen', edgecolor = "#6A9662")
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
plt.figure(figsize = (10,8)) #figÃ¼r boyutunu ayarlama
x = np.random.normal(size = 2000) #normal daÄÄ±lÄ±mlÄ± random sayÄ± oluÅturma
plt.hist(x, bins = 10, color = 'yellowgreen', edgecolor = "#6A9662") #histogram fonksiyonu
plt.gca().set(title = 'Histogram', ylabel = 'Frequency') #get current axes
plt.show()
```

```python
#Multiple Histogram
plt.figure(figsize = (8, 11))
x = np.random.normal(-4, 1, size = 800)
y = np.random.normal(0, 1.5, size = 800)
z = np.random.normal(3.5, 1, size = 800)
plt.hist(x, bins = 30, color = 'yellowgreen', alpha = 0.6)
plt.hist(y, bins = 30, color = '#FF8F00', alpha = 0.6)
plt.hist(z, bins = 30, color = 'blue', alpha = 0.6)
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
#Histogram kullanarak Cumulative Distribution Function (CDF) Ã§izdirme 
```

```python
plt.figure(figsize = (10,8))
x = np.random.rand(2000) #toplam 2000 adet 0-1 arasÄ±nda rasyonel sayÄ± oluÅturuyor.
plt.hist(x, bins = 30, color = '#FFA41B', edgecolor = '#639A67', cumulative = True)
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
col1 = np.linspace(0, 10, 1000)
col2 = np.sin(col1)
df = pd.DataFrame({'C1' : col1, 'C2' : col2})
df.head()
```

```python
canada = pd.read_csv('mydata/canada1.csv', delimiter = ';')
canada.head()
```

```python
canada.drop(columns = ['AREA', 'DEV', 'DevName', 'REG', 'Type', 'Coverage', 'AreaName', 'RegName'], inplace = True)
canada.head()
```

```python
canada.rename(columns = {'OdName' : 'Country'}, inplace = True)
canada.set_index(canada.Country, inplace = True)
canada.head()
```

```python
canada.index.name = None
canada.head()
```

```python
del canada['Country']
canada.head()
```

```python
canada = canada.transpose()
canada.head()
```

```python
plt.figure(figsize = (14,6))
plt.title('Immigrants from TÃ¼rkiye', fontsize = 14)
sns.set(rc = {"axes.facecolor" : '#283747', "axes.grid" : False, "xtick.labelsize" : 10, 'ytick.labelsize' : 10})
plt.xticks(rotation = 45)
sns.lineplot(x = canada.index.values, y = canada['Turkey'])
plt.show()
```

```python
plt.figure(figsize = (14,6))
plt.title('Immigrants', fontsize = 14)
sns.set(rc = {"axes.facecolor" : '#283747', "axes.grid" : False, "xtick.labelsize" : 10, 'ytick.labelsize' : 10})
plt.xticks(rotation = 45)
sns.lineplot(x = canada.index.values, y = canada['Ukraine'], label = 'Ukraine')
sns.lineplot(x = canada.index.values, y = canada['Viet Nam'], label = 'Vietnam')
sns.lineplot(x = canada.index.values, y = canada['Syrian Arab Republic'], label = 'Syria')
sns.lineplot(x = canada.index.values, y = canada['Poland'], label = 'Poland')
plt.legend(facecolor = 'grey', fontsize = 'large', edgecolor = 'black', shadow = True)
plt.show()
```
